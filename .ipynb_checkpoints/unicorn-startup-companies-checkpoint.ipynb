```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('unicorn-startup-companies.ipynb')
```

```python
df.info()
```

# Which countries have the most unicorn startups?

```python
import plotly.express as px
country_counts = df['Country/countries'].value_counts()
top_countries = country_counts.head(10)
fig = px.bar(top_countries, x=top_countries.values, y=top_countries.index, orientation='h')
fig.update_layout(title='Top 10 Countries with the Most Unicorn Startups', xaxis_title='Number of Unicorn Startups', yaxis_title='Country')
fig.show()
```

# visualize the number of unicorn startups by region 

```python
regions = {'North America': ['US', 'Canada', 'Mexico'],
           'Europe': ['UK', 'Germany', 'France', 'Israel', 'Sweden', 'Netherlands', 'Ireland', 'Switzerland', 'Norway', 'Spain', 'Belgium'],
           'Asia': ['China', 'India', 'South Korea', 'Singapore', 'Indonesia', 'Japan', 'Hong Kong'],
           'South America': ['Brazil'],
           'Australia': ['Australia']}

df['Region'] = df['Country/countries'].apply(lambda x: next((k for k,v in regions.items() if x in v), None))
region_counts = df.groupby('Region').size().fillna(0)
fig = px.bar(region_counts, x=region_counts.index, y=region_counts.values, orientation='v', color=region_counts.index)
fig.update_layout(title='Unicorn Startups by Region', xaxis_title='Region', yaxis_title='Number of Unicorn Startups')
fig.show()
```

```python

country_counts = df.groupby(['Country/countries', 'Valuation date']).size().reset_index(name='count')
fig = px.line(country_counts, x='Valuation date', y='count', color='Country/countries')
fig.update_layout(title='Unicorn Startups by Country and Year', xaxis_title='Valuation Date', yaxis_title='Number of Unicorn Startups')
fig.show()
```
